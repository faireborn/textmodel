```python
freq = defaultdict (int)
with open ('data/alice.txt', 'r') as f:
    for line in f:
        s = line.rstrip('\n')
        L = len(s)
        s += '$'
        for i in range(L):
            b = s[i:i+2]
            freq[b] += 1
```

```python
for (b,n) in sorted (freq.items(), key=lambda x: x[1], reverse=True):
    print ('%s = %d' % (b,n))
print ('total %d bigrams.' % (sum(list(freq.values()))))
```

```python
%run bin/bigram-plot.py data/alice.txt
```

```python
p = {}
N = sum(list(freq.values()))
for (b,n) in freq.items():
    x = b[0]; y = b[1]
    if not (x in p):
        p[x] = {}
    p[x][y] = n / N
```

```python
p
```

```python
for x in p:
    s = 0
    for y in p[x]:
        s += p[x][y]
    print ('p(%s) = %f' % (x,s))
```

```python
p0  # 前でユニグラムから計算した値と比較してみる
```

```python
p0 = {}; chars = p.keys()
for c in chars:
    p0[c] = sum(list(p[c].values()))
c = {}
for x in chars:
    c[x] = {}
    for y in chars:
        if y in p[x]:
            c[x][y] = p[x][y] / p0[x]
        else:
            c[x][y] = 0    
```
